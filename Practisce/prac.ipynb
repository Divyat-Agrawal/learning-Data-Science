```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,2*np.pi,0.1)
y= np.sin(x)
# z= np.cos(x)
plt.plot(x,y,color = "blue" , marker = "o", linestyle = "dotted")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend(['sinx'])
plt.show()
```

```python
# DENSITY PLOT

import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
data = np.random.randn(1000)  # Generate 1000 random data points from a standard normal distribution

# Create a histogram
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)  # Create a subplot for the histogram

# You can adjust the number of bins, color, and transparency
plt.hist(data, bins=20, color='skyblue', alpha=0.7, edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram")

# Create a density plot
plt.subplot(1, 2, 2)  # Create a subplot for the density plot

# You can adjust the bandwidth and color
plt.hist(data, bins=100, density=True, color='salmon', alpha=0.7, edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Density Plot")

plt.tight_layout()  # Adjust spacing for subplots
plt.show()


```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x = np.random.rand(50)  # 50 random x-coordinates
y = np.random.rand(50)  # 50 random y-coordinates
colors = np.random.rand(50)  # Random color values
sizes = 100 * np.random.rand(50)  # Random sizes for points

# Create a scatter plot
plt.scatter(x, y, c=colors, s=sizes, cmap='viridis', alpha=0.7, edgecolors='k')

# Add labels and a colorbar
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot with Different Parameters")

# Add a color bar to show the relationship between color and data values
plt.colorbar()

# Display the plot
plt.show()
```
